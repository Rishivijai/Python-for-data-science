# Data Visualization
Is the presentation of data in a graphical or pictorial format. It involves creating visual representations of data to identify patterns, trends, and insights that may not be apparent in raw data.

# Matplotlib 

Matplotlib is a comprehensive library for  creating static, animated, and interactive visulization in Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# let's add more to it
# Let's create a bar graph
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
ax.bar(x,y)
ax.set_xlabel("Product")
ax.set_ylabel("Unit Sold")
ax.set_title("Units sold for each product")
ax.grid(True) # shows grid
```

```python
# let's add more to it
# Let's create a bar graph
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
bar_color = ["Red","pink","Blue"]
ax.bar(x,y,color=bar_color)
ax.set_xlabel("Product")
ax.set_ylabel("Unit Sold")
ax.set_title("Units sold for each product")
ax.grid(True) # shows grid
```

# Histogram 

```python
fig,ax = plt .subplots(figsize=(10,5))
np.random.seed(12)
a = np.random.randint(100,size = 80) # generate 80 random integers between 0 and 100(exclusive)
ax.hist(a,bins = 5)
ax.set_title("Random Generation")
```

# Scatter plot 

```python
x = np.linspace(1,20)
y = [i**2 for i in x]
y1 = [i**3 for i in x]

fig,ax = plt.subplots(figsize=(5,5))
ax.scatter(x,y,alpha=0.7,color="red",marker="*")
ax.scatter(x,y1,alpha=0.8,color="blue",marker="o")
ax.legend(["**2","**3"])
ax.grid(True)
ax.set_title("Scatter Plot")
ax.set_xlabel
```

```python
# Pie chart
company = ["Samsung","Redmi","OnePlus","Oppo"]
size = [300,20,15,35]
fig,ax = plt.subplots(figsize=(5,5))
ax.pie(size,labels=company,autopct="%1.2f%%")
ax.set_title("Market Share")
```
